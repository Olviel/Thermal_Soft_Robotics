```python
from scipy.integrate import solve_ivp
import numpy as np

def system(t, y, params):
    var1, var2 = y
    const1, const2 = params
    dydt = [const1 * var1 - const2 * var2, var1 + var2]
    return dydt

y0 = [1, 0]
params = [2, 3]
t_span = [0, 10]  # start and end time
array = np.linspace(t_span[0],t_span[1],2)

sol = solve_ivp(system, t_span, y0, args=(params,), t_eval=array,dense_output=True)

t=sol.t
y=sol.y
# = np.linspace(0, 10, 1000)
#y = sol.sol(t)
```
