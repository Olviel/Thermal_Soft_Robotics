```python
import graphviz

# Create the main Digraph with Left to Right direction
dot = graphviz.Digraph('unix', graph_attr={'rankdir': 'TB','splines':'true'})


# Add nodes
dot.node('A','$P_{amb}$, $T_{amb}$, $n_{air}$, n_{novec}^l,n_{novec}^g,\\ Valve= open', shape='box') 
dot.node('B', 'Update P_{sun}, Q_{rad}, Q_{con},Q_{evap}')

dot.node('D', '$T>T_{boil}$')
dot.node('E')
dot.node('F')
dot.node('G')



# Add edges to align nodes vertically and connect C, C1, C2
#dot.edges(['AB', 'BC', 'CD', 'DE', 'EF', 'FG', 'GH'])
dot.edge('A', 'B' )
dot.edge('B', 'C')
dot.edge('C', 'D',color='crimson')
dot.edge('D', 'E',color='crimson')
dot.edge('E', 'F',color='crimson')
dot.edge('F', 'G',color='crimson')


# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('C')
    s.node('C1')
    s.node('C2')

dot.edge('C', 'C1',color='forestgreen')
dot.edge('C1', 'D',color='crimson')
dot.edge('C1', 'C2',color='forestgreen')
dot.edge('C2','D')


# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('D')
    s.node('D1')
    s.node('D2')

dot.edge('D', 'D1',color='forestgreen')
dot.edge('D1', 'E',color='crimson')
dot.edge('D1', 'D2',color='forestgreen')
dot.edge('D2','E',color='crimson')


# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('E')
    s.node('E1')

dot.edge('E', 'E1',color='forestgreen')
dot.edge('E1', 'F')

# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('F')
    s.node('F1')
    s.node('F2')

dot.edge('F', 'F1',color='forestgreen')
dot.edge('F1', 'F2',color='forestgreen')
dot.edge('F1', 'G',color='crimson')
dot.edge('F2','G')


dot.edge('G','C',color='crimson')
dot.edge('G','H',color='forestgreen')

dot.node('H', shape='box') 

dot.node('True',shape='plaintext',fontcolor='forestgreen')
dot.node('False', shape='plaintext',fontcolor='crimson')
#dot.node('Continue', shape='plaintext',fontcolor='Black')








# Display the graph
dot
```

```python
# Export the graph to an SVG file
dot.render('output_graph', format='svg', cleanup=True)
```
