```python
import graphviz

# Create the main Digraph with Left to Right direction
dot = graphviz.Digraph('unix', graph_attr={'rankdir': 'TB'})


# Add nodes
dot.node('A','$P_{amb}$', shape='box') 
dot.node('B')
dot.node('D')
dot.node('E')
dot.node('F')
dot.node('G')
dot.node('H', shape='box') 

# Add edges to align nodes vertically and connect C, C1, C2
dot.edges(['AB', 'BC', 'CD', 'DE', 'EF', 'FG', 'GH'])


# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('C')
    s.node('C1')
    s.node('C2')

dot.edge('C', 'C1',label='True')
dot.edge('C1', 'D',label='F')
dot.edge('C1', 'C2',label='True')
dot.edge('C2','D')


# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('D')
    s.node('D1')
    s.node('D2')

dot.edge('D', 'D1')
dot.edge('D1', 'E')
dot.edge('D1', 'D2')
dot.edge('D2','E')


# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('E')
    s.node('E1')

dot.edge('E', 'E1')
dot.edge('E1', 'F')

# Create a subgraph for nodes that should be on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    # Add nodes to the subgraph
    s.node('F')
    s.node('F1')
    s.node('F2')

dot.edge('F', 'F1')
dot.edge('F1', 'F2')
dot.edge('F1', 'G',label='False')
dot.edge('F2','G')


dot.edge('G','C')










# Display the graph
dot
```

```python
# Export the graph to an SVG file
dot.render('output_graph', format='svg', cleanup=True)
```
