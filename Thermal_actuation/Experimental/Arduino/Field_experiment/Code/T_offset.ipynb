```python
import numpy as np
import matplotlib.pyplot as plt

# Data
TC = np.array([24,29,42,47,49,54,63,67,72,76])
Real = np.array([26,43,48,54,61,65,74,80,85,91])

# Perform linear fit
m, c = np.polyfit(TC, Real, 1)

# Perform second-order polynomial regression
a, b, c2 = np.polyfit(TC, Real, 2)

# Create range of TC values for plotting
tc_range = np.linspace(min(TC), max(TC), 100)

# Compute the corresponding Real values
linear_fit = m * tc_range + c
second_order_fit = a * tc_range**2 + b * tc_range + c2

# Plot original data
plt.scatter(TC, Real, color='blue', label='Original Data')

# Plot linear fit
plt.plot(tc_range, linear_fit, color='red', label='Linear Fit')

# Plot second-order fit
#plt.plot(tc_range, second_order_fit, color='green', label='Second-Order Fit')

range_x = np.linspace(0,100,10)
range_y = np.linspace(0,100,10)
plt.plot(range_x,range_y, label='Ideal')

plt.xlabel('TC')
plt.ylabel('Real')
plt.legend()
plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
print(m,c)
```
