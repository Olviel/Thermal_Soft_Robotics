```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
P0 = 200000  # Initial tank pressure, Pa
V = 0.003  # Tank volume, m^3
P_ambient = 101325  # Ambient pressure, Pa
R = 50000000  # Flow resistance, 

# Time step and number of steps
dt = 0.01  # Time step, s
n_steps = 2000  # Number of steps

# Arrays to store the results
time = np.zeros(n_steps)
P = np.zeros(n_steps)
Q = np.zeros(n_steps)

# Initial conditions
P[0] = P0

# Simulation loop
for i in range(1, n_steps):
    Q[i-1] = max((P[i-1] - P_ambient) / R, 0)  # Flow rate (ensure it's not negative)
    P[i] = P[i-1] - Q[i-1] * dt * P[i-1] / V  # Update pressure
    time[i] = i * dt  # Update time

# Plot results
plt.figure()

plt.subplot(211)
plt.plot(time, P / 1e5)  # Pressure in bar
plt.xlabel('Time [s]')
plt.ylabel('Pressure [bar]')

plt.subplot(212)
plt.plot(time[:-1], Q[:-1] * 1e6)  # Flow rate in L/s
plt.xlabel('Time [s]')
plt.ylabel('Flow rate [L/s]')

plt.tight_layout()
plt.show()
```
